#  Model Training: KNN algorithm

1. Creating a pivot table to show user ratings for each book.
Each row represents a book title, each column represents a user ID,
and the values indicate the ratings given by users (scaled from 0 to 1).
2. clustering: We're going to find users who gave the same ratings to the same books. Then, we'll focus on recommending books they rated because they likely have similar interests.


```python
books_pivot=rating_count.pivot_table(columns='user_id',index='title',values='rating')
```

```python
books_pivot.head()
```

```python
books_pivot.shape
```

```python
books_pivot.fillna(0,inplace=True)
```

```python
books_pivot.head(2)
```

```python
# Importing the csr_matrix function from the scipy.sparse module.
# We'll use this function to create Compressed Sparse Row (CSR) matrices, 
# which are optimized for datasets with many zero values.
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(books_pivot)
```

```python
book_sparse
```

```python
#clustering
#distance calculation
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
# Finding the nearest neighbors for a specific book and printing the suggestions.
distance,suggestion=model.kneighbors(books_pivot.iloc[254,:].values.reshape(1,-1),n_neighbors=5)
```
